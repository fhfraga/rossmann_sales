## FUNÇÕES DE APOIO

```python
# função para achar v de cramer
def cramer_v( x, y ):
    cm = pd.crosstab( x, y ).values
    n = cm.sum()
    r, k = cm.shape
    
    chi2 = stats.chi2_contingency( cm )[0]
    chi2corr = max( 0, chi2 - (k-1)*(r-1)/(n-1) )
    
    kcorr = k - (k-1)**2/(n-1)
    rcorr = r - (r-1)**2/(n-1)
    
    return np.sqrt( (chi2corr/n) / ( min( kcorr-1, rcorr-1 ) ) )

def mean_absolute_percentage_error_2(y, yhat):
    return np.mean(np.abs((y - yhat) / y))
    

def ml_error(model_name, y, yhat):
    mae = mean_absolute_error(y, yhat)
    mape = mean_absolute_percentage_error(y, yhat)
    rmse = np.sqrt(mean_squared_error(y, yhat))
    
    return pd.DataFrame({'Nome do modelo': model_name, 'MAE': mae, 'MAPE': mape,
                         'RMSE': rmse}, index=[0])


# função para cross validation
def cross_validation( x_training, kfold, model_name, model, verbose=False ):
    mae_list = []
    mape_list = []
    rmse_list = []
    for k in reversed( range( 1, kfold+1 ) ):
        if verbose:
            print( '\nKFold Number: {}'.format( k ) )
        # start and end date for validation 
        validation_start_date = x_training['date'].max() - datetime.timedelta( days=k*6*7)
        validation_end_date = x_training['date'].max() - datetime.timedelta( days=(k-1)*6*7)

        # filtering dataset
        training = x_training[x_training['date'] < validation_start_date]
        validation = x_training[(x_training['date'] >= validation_start_date) & (x_training['date'] <= validation_end_date)]

        # training and validation dataset
        # training
        xtraining = training.drop( ['date', 'sales'], axis=1 ) 
        ytraining = training['sales']

        # validation
        xvalidation = validation.drop( ['date', 'sales'], axis=1 )
        yvalidation = validation['sales']

        # model
        m = model.fit( xtraining, ytraining )

        # prediction
        yhat = m.predict( xvalidation )

        # performance
        m_result = ml_error( model_name,  yvalidation, yhat )

        # store performance of each kfold iteration
        mae_list.append(  m_result['MAE'] )
        mape_list.append( m_result['MAPE'] )
        rmse_list.append( m_result['RMSE'] )

    return pd.DataFrame( {'Nome do modelo': model_name,
                          'MAE CV': np.round( np.mean( mae_list ), 2 ).astype( str ) + ' +/- ' + np.round( np.std( mae_list ), 2 ).astype( str ),
                          'MAPE CV': np.round( np.mean( mape_list ), 2 ).astype( str ) + ' +/- ' + np.round( np.std( mape_list ), 2 ).astype( str ),
                          'RMSE CV': np.round( np.mean( rmse_list ), 2 ).astype( str ) + ' +/- ' + np.round( np.std( rmse_list ), 2 ).astype( str ) }, index=[0] )

def mean_percentage_error(y, yhat):
    return np.mean((y - yhat) / y)

# parâmetros para os gráficos
params = {
    'lines.linewidth': 2.0,
    'axes.labelsize': 16,
    'axes.titlesize': 16, 
    'axes.spines.top': False,
    'axes.spines.right': False,
    'xtick.labelsize': 16,
    'ytick.labelsize': 16, 
    'figure.autolayout': True,
    'figure.titlesize': 24,
    'figure.figsize': (14, 8),
    'legend.shadow': False,
    'legend.fontsize': 14,
}

plt.rcParams.update(params)

# ignorando as mensagem de aviso
warnings.filterwarnings("ignore")
```

Algumas formas de preencher esses valores podem ser:
1. **Excluir as entradas (linhas)**: É um método rápido, porém jogar uma quantidade grande de dados fora pode prejudicar o modelo preditivo que será criado.
2. **Algum método estatístico (média, mediano, moda,...)**: Um método muito utilizado, porém deve ser ver com cautela para não escolher o errado. Por exemplo, usar a média quando se tem um outlier muito distante dos valores, pode prejudicar o modelo preditivo ou até mesmo os gráfico criados.
3. **Entendimento de negócio**: Estuda-se o negócio para entender o porque dos valores ausentes, pois os mesmo pode ser uma falta de informação ou mesmo um padrão de negócio. Entendendo a regra é possível substituir esses valores de NA's por valores mais próximos da realidade.

    * `Método possível de resolver`: Copiar os valores NA's da data que está na linha (data de venda). Porém, por mais ilógico que seja de inicio, é apenas uma forma inicial de tratar os dados.

* **competition_open_since_year**: Um ano aproximado que o competidor mais próximo foi aberto.

    * `Método possível de resolver`: Copiar os valores NA's o ano da data que está na linha (data de venda). Porém, por mais ilógico que seja de inicio, é apenas uma forma inicial de tratar os dados.

## ESTATÍSTICA DESCRITIVA

A descrição estatística dos dados serve para:
1. Ganhar conhecimento de negócio
2. Possível detecção de alguns erros

Há duas grandes métricas em estatística descritiva: métricas de dispersão e as métricas de tendências central.

* **Métricas de tendência central**: Resumem os dados em um único número. Ex.: média, mediana
* **Métricas de dispersão**: Dizem o qual está concentrado ou disperso em relação a média. Ex.: Variância, desvio padrão.
    * `Desvio-Padrão`: É a distância dos pontos em relação a média;
    * `range`: valor entre mínimo e máximo;
    * `Assimetria (skew)`: Diz como é a deformação da distribuição em relação à normal. Uma deformação maior para esquerda é uma skew positiva; uma deformação maior para direita é uma skew negativa;
    * `Curtose`: Diz sobre a concentração dos dados. Quanto maior a curtose positiva, maior a concentração, dando picos altos; curtose com valores tendendo a negativos tem picos menores e são maios dispersas.

# PASSO 04 - ANÁLISE EXPLORATÓRIA DOS DADOS

A análise exploratória dos dados serve para medir o impacto das variáveis em relação a variável resposta e muitas vezes tentar quantificar esse impacto.

**Quais os objetivos da EDA(análise exploratória dos dados)?**

1. **Ganhar experiência de negócio**:
    * Entender como o negócio funciona, o comportamento que o negócio tem através dos dados.

2. **Validar hipóteses de negócio (Insights)**:
    * Gerar insights: 
        1. Gerar surpresa: Fornecer informação que as pessoas desconhecem.
        2. Chocar uma crença da pessoa: Mostrar que uma informações empíricas, estão erradas quando verificamos os dados.
        
3. **Perceber variáveis que são importantes para o modelo**: 
    * Quando se faz a EDA, ganha-se sensibilidade de quais variáveis impactam o fenômeno, criando um modelo com comportamento melhor.

A Análise Exploratória de Dados é constituída de 3 tipos de análises.

1. **Análise Univariada**: Avalia-se apenas uma única variável.
    * Como é essa variável? (Min, Max, Distribuição, Range, ...)
    
2. **Análise Bivariada**: Avalia-se o conjunto de dados e olha qual o impacto de determinada variável na variável resposta.
    * Como a variável impacta na resposta? (Correlação, Validação de Hipóteses, ...)
    
3. **Análise Multivariada**: Avalia-se o conjunto de dados e olha qual o impacto de múltiplas variáveis na variável resposta e entre elas.
    * Como as variáveis se relacionam? (Qual a correlação entre elas?)

## ANÁLISE MULTIVARIADA

Os algoritmos de machine learning seguem algumas premissas, uma delas é a [Navalha de Ockham](https://pt.wikipedia.org/wiki/Navalha_de_Ockham), sendo uma das teorias que garante o aprendizado dos modelos. Essa teoria fala que se você tiver vários modelos para escolher, deve-se sempre escolher o de menor complexidade, pois ele generaliza o aprendizado melhor.

Existem várias formas de tornar um algoritmo mais complexo, uma delas é a dimensionalidade do dataset. Dimensionalidade é o número de variáveis (colunas) do dataset. Quanto mais variáveis, maior o dataset e mais complexo o modelo. Para diminuir a dimensionalidade, pode-se retirar variáveis de seu dataset.

**Mas como se tira variáveis de um dataset? Qual critério utilizar?**

Bom, um dos critérios usados é o quanto de informação cada uma das colunas carrega. Exemplo, imagine que tem duas variáveis no dataset carregando a mesma informação. Nesse caso é possível retirar uma das variáveis pois não se perderá a informação dela, pois a outra contém a informação.

**E como encontrar quais variáveis carregam a mesma quantidade de informação?**

Para encontrar isso usa-se um método de algebra linear chamado de **vetores linearmente dependentes**. Onde se tiver dois vetores, ou duas variáveis, que são linearmente dependentes, pode-se retirar uma que o conteúdo de informação se mantém. Uma das formas de ver isso é pela correlação da variáveis. Onde se uma variável for muito correlacionada com a outra, pode-se retirar uma que o conteúdo da informação se manterá.

A **Análise Multivariada** te ajuda a olhar quais variáveis que são correlacionadas e portanto podem ser retiradas para diminuir a dimensionalidade do dataset e diminuir a complexidade do algoritmo.

Para as variáveis categóricas usaremos o método de correlação de [Cramér's V](https://en.wikipedia.org/wiki/Cram%C3%A9r%27s_V). Esse método vai apenas de 0 até 1. Onde quanto mais próximo de 0 menor a correlação, e quanto mais próximo do 1, maior a correlação.

Sua fórmula é:

$$V = \sqrt{\frac{\sigma^2}{min(k-1, r-1)}} = \sqrt{\frac{X^2/n}{min(k-1, r-1)}}$$

onde:
* $\sigma$:é o coeficiente phi.
* $X²$ é derivado do teste qui-quadrado de Pearson
* $n$ é o total geral de observações e
* $k$ sendo o número de colunas.
* $r$ sendo o número de linhas.

**OBS.:** Lembrando que a função contendo a fórmula já está em `funções de apoio`.

```python
# calculo do v de cramer
cat_att_1 = cramer_v(cat_attributes['state_holiday'], cat_attributes['state_holiday'])
cat_att_2 = cramer_v(cat_attributes['state_holiday'], cat_attributes['store_type'])
cat_att_3 = cramer_v(cat_attributes['state_holiday'], cat_attributes['assortment'])

cat_att_4 = cramer_v(cat_attributes['store_type'], cat_attributes['state_holiday'])
cat_att_5 = cramer_v(cat_attributes['store_type'], cat_attributes['store_type'])
cat_att_6 = cramer_v(cat_attributes['store_type'], cat_attributes['assortment'])

cat_att_7 = cramer_v(cat_attributes['assortment'], cat_attributes['state_holiday'])
cat_att_8 = cramer_v(cat_attributes['assortment'], cat_attributes['store_type'])
cat_att_9 = cramer_v(cat_attributes['assortment'], cat_attributes['assortment'])

# dataset final
df_cat_att = pd.DataFrame({'state_holiday': [cat_att_1, cat_att_2, cat_att_3],
              'store_type': [cat_att_4, cat_att_5, cat_att_6],
              'assortment': [cat_att_7, cat_att_8, cat_att_9]})

df_cat_att = df_cat_att.set_index(df_cat_att.columns)
df_cat_att
```

## RESCALING
dos modos de fazer o rescaling é usando o **Min-Max Scaler** e o **Robust Scaler**

**Usos**:
* **Min-Max Scaler**: Utilizado para variáveis que não possuem outliers muito fortes.
    
    $$x_{new} = \frac{X_i - \mu}{x_{max} - x_{min}}$$

  * onde:
    * $X_{new}:$ variável nova (valores novos da coluna)
    * $X_i$: variável original (valores da coluna)
    * $\mu$: média
    * $x_{max} - x_{min}$: range (válor máximo - válor mínimo)
    
    
* **Robust Scaler**: Utilizado para variáveis que possuem outliers muito fortes.

    $$x_{new} = \frac{X_i - Q_1(x)}{Q_3(x) - Q_1(x)}$$

    * onde:
        * $X_{new}:$ variável nova (valores novos da coluna)
        * $X_i$: variável original (valores da coluna)
        * $Q_1(x)$: Primeiro quartil da variável
        * $Q_3(x)$: Terceiro quartil da variável

### ENCODING

Encoding é o método de transformar variáveis categóricas em numéricas pra que os algoritmos de aprendizado de máquina possam ser treinados. Nesse projeto utilizaremos as transformações `One Hot Encoding`, `Label Enconder` e `Ordinal Encoding`

$$\textbf{One Hot Encoding}$$

Transforma cada linha da variável categórica em uma coluna, onde se na linha onde era a variável passará a ter o valor de 1 (True), se não será 0 (False).

$$\textbf{Label Encoding}$$

Troca as variáveis categóricas por números, que podem ser aleatórios, porém se o mesmo valor se repetir, deve ser dado o mesmo número anteriormente dado.
* **Uso**: Quando se tem variáveis que são apenas nomes, que não tem relação explicita ou nenhuma relação.

$$\textbf{Ordinal Encoding}$$

Troca as variáveis categóricas por números em uma hierarquia, uma ordem de maior ou menor.
* **Uso**: Quando se tem variáveis que possuem algum tipo de ordem, hierarquia, como temperatura, altura, nome em ordem alfabética, etc.

```python
# state_holiday - One Hot Enconding
df5 = pd.get_dummies(df5, prefix=['state_holiday'], columns=['state_holiday'])
```

### TIPOS DE TRANSFORMAÇÕES

Há dois tipos de conjuntos de transformações
1. Transformações de grandeza
2. Transformações de natureza

$$\textbf{Transformações de Grandeza}$$

Tem o objetivo de trazer sua variável resposta com um distribuição mais próxima possível de uma distribuição normal.

Exemplo: Os algoritmos de machine learning foram criados sobre a premissa que os dados são normalmente distribuídos. Então quanto mais próximo os dados dão match com as premissas, maior a acurácia.

AS transformações de grandeza são:
1. Logaritma
2. Box-Cox
3. Cube-Root
4. Square-Root
5. Seno e Cosseno

$$\textbf{Transformações de Natureza}$$

Tem o objetivo de trazer a natureza real dos seus dados dentro do conjunto de dados.

Exemplo: Uma variável é o mês, dando o número de 1 a 12, de janeiro a fevereiro. Janeiro está próximo de fevereiro, já que 1 está próximo do 2, porém janeiro está longe de dezembro, já que 1 está distante de 12. Porém quando vira o ano, dezembro, que é 12, fica muito próximo de janeiro do outro ano que é 1. Porém se levar em consideração apenas os número, isso não daria certo, já que tem que se pensar ciclicamente. Por isso a transformação de natureza.

<p style="text-align: center;"> <b> Transformações de Natureza Cíclica</b> </p>

A forma de transformar é utilizando o círculo trigonométrico. Pois como dito acima, tem variáveis que numericamente não demonstrariam uma relação próxima que na verdade elas têm.

```python
# day of week
df5['day_of_week_sin'] = df5['day_of_week'].apply(lambda x: np.sin(x * (2. * np.pi/7)))
df5['day_of_week_cos'] = df5['day_of_week'].apply(lambda x: np.cos(x * (2. * np.pi/7)))

# month
df5['month_sin'] = df5['month'].apply(lambda x: np.sin(x * (2. * np.pi/12)))
df5['month_cos'] = df5['month'].apply(lambda x: np.cos(x * (2. * np.pi/12)))

# day
df5['day_sin'] = df5['day'].apply(lambda x: np.sin(x * (2. * np.pi/30)))
df5['day_cos'] = df5['day'].apply(lambda x: np.cos(x * (2. * np.pi/30)))

# week of year
df5['week_of_year_sin'] = df5['week_of_year'].apply(lambda x: np.sin(x * (2. * np.pi/52)))
df5['week_of_year_cos'] = df5['week_of_year'].apply(lambda x: np.cos(x * (2. * np.pi/52)))
```

O princípio da Navalha de Occam, que diz:

**"A explicação mais simples sobre um fenômeno observado, deveria prevalecer sobre explicações mais complexas"**

Um modelo é uma simplificação, uma descrição de um fenômeno que pode ser replicado.

Pela navalha de Occam, escolheremos os modelos mais simples que explicam o fenômeno. Como visto anteriormente, uma meio de aumentar a simplificação de uma modelo é ele possuir menos variáveis, porém, pela navalha de Occam, modelos que generalizam o fenômeno e são simples, devem ser os escolhidos, de nada adianta ter um modelo simples mas que não explica o fenômeno observado. 

Pensemos como exemplo um carro, se tivermos um modelo que tem poucas variáveis, mas tem por exemplo bluetooth e câmbio automático como variáveis, não temos um modelo que generaliza bem sobre carros. Agora um modelo com várias variáveis, mas que contém rodas, lanternas, retrovisor, portas, possui um generalização melhor, mesmo tendo uma dimensionalidade maior.

# PASSO 08 - HYPERPARAMETER FINE TUNING

**Fine tuning** é a forma de encontrar o conjunto de parâmetros que maximiza o aprendizado do modelo.

Cada modelo tem parâmetros e o conjunto de parâmetros é chamado de hiperparâmetro. Cada parâmetro pode assumir uma infinidade de valores. Isso faz com que seja díficil decidir quais valores serão os parâmetros, dai entra o fine tuning. 

Há 3 estratégias para o fine tuning.
* Random Search
    1. Define os valores para cada um dos hiperparâmetros aleatoriamente.
* Grid Search
    1. Define todas as combinações possíveis de valores que os hiperparâmetros podem assumir.
* Bayesian Search
    1. Define os valores para os hiperparâmetros segundo a teoria de Bayes.

```python
param = {
    'n_estimators': [15, 17, 25, 30, 35],
    'eta': [0.01, 0.03],
    'max_depth': [3, 5, 9],
    'subsample': [0.1, 0.5, 0.7],
    'colsample_bytree': [0.3, 0.7, 0.9],
    'min_child_weight': [3, 8, 15]
}

MAX_EVAL = 10 # tentar no mínimo 10 iterações
```

```python
final_result = pd.DataFrame()

for i in range(MAX_EVAL):
    # escolher os valores para os parâmetros aleatoriamente
    hp = {k: random.sample(v, 1)[0] for k, v in param.items()}
    print(hp)

    # modelo
    model_xgb = xgb.XGBRegressor(objective='reg:squarederror', 
                                 n_estimators=hp['n_estimators'],
                                 eta=hp['eta'],
                                 max_depth=hp['max_depth'],
                                 subsample=hp['subsample'],
                                 colsample_bytree=hp['colsample_bytree'],
                                 min_child_weight=hp['min_child_weight'])
    
    # perfomance
    result = cross_validation(x_training, 2,'XGBoost Regressor',  model_xgb, verbose=False)
    final_result = pd.concat([final_result, result])
    
final_result
```

# PASSO 09 - INTERPRETAÇÃO E TRADUÇÃO DO ERRO

Entender a perfomance do Modelo e traduzir para a empresa o quanto esse modelo irá  trazer de benefício para empresa.

Esse passo pode ser dividido em:

1. **Perfomance do Modelo de Regressão**
    1. **MAE (Mean Absolute Error - Erro Médio Absoluto)**
        * É a diferença entre o valor real e o valor predito. Possui facil entendimento, atribui peso igual para todos, sendo ruim para uso com outliers.
    2. **MAPE (Mean Absolute Percentage Error - Erro Médio Absoluto em Porcentagem)**
        * É a diferença entre o valor real e o valor predito, sendo esse valor colocado em módulo para ter seu valor absoluto e posteriormente dividido pelo valor real, dando um valor entre 0 e 1, multiplicando por 100 dá o valor em porcentagem. Possui fácil entendimento, mostra o quão longe a  predição está do valor real, contudo, não pode ser utilizado se o a variável resposta possuir em algumas de suas linhas o 0. 
    3. **RMSE (Root Mean Squared Erro - Raiz Quadrada do Erro Médio**
        * É a diferença entre a predição real e a predita, sendo essa diferença elevada ao quadrado, todas as diferenças de toda a coluna são somadas e divida pelo número de amostras. Desse valor é retirada a raiz quadrada. Ele atribui maior peso a erros maiores, é sensível na presença de outliers.
    4. **MPE (Mean Percentage Error - Erro Médio em Porcentagem)**
        * Muito parecido com o MAPE, sendo a principal diferença a ausência do módulo. Isso se deve a esse erro querer ver se o modelo está subestimando (modelo prevê abaixo do que é real) ou superestimando (modelo prevê acima do que é real). Não é utilizado para medir a perfomance do modelo, pois os erros podem se cancelar.
    
2. **Perfomance do Modelo sobre o Negócio**: Pegar a acurácia de um modelo e traduzir para um valor que é trabalhado pela empresa (dinheiro, número de novos usuários, etc...)
    1. **Comparação com o Status quo (estado das coisas)**
        * É comparar o estado do modelo com a predição feita por outra equipe, seja de negócios, seja por um gerente, sendo essa predição feita pela experiência, por alguma método.
    2. **Incremento de Valor (Aumento de receita, aumento de usuário, velocidade de algo acontecer)**
        * O quanto ele gera de valor para a empresa

## COMO USAR OS ERROS

1. **Perfomance do Modelo de Regressão**
    1. RMSE e MPE
        * Métricas para melhoria de performance do Modelo
        
2. **Performance do Modelo sobre o Negócio**
    1. MAE, MAPE e MPE
        * Métricas para reportar ao Time de Negócio
